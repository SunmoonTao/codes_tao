```bash
### Setting Paths

Input_DIR=/Users/ltao/Projects/glioma/exp/190405/fastqs
hg19=/Users/ltao/ref/human/human_g1k_v37.fasta
```

```bash
ls $Input_DIR
```

```bash
# bwa mapping ---> converted to bam by samtools:

for a in `ls  $Input_DIR/*MERGED`
do
    TMP_NAME=`basename $a`
	echo $TMP_NAME
    bwa mem $hg19 $a | samtools view -bS - > $Input_DIR/$TMP_NAME.bam
    echo 'next'    
done
```

```bash
## Sort and index BAM by samtools
for BAM  in `ls  $Input_DIR/*.bam`
do
    TMP_NAME=`basename $BAM`
# 	echo $TMP_NAME
	SUB_NAME=`echo $TMP_NAME |awk -F '.bam' '{print $1}'`
	echo $SUB_NAME
    mkdir -p $Input_DIR/tmp
    echo 'sorting' 
    echo $SUB_NAME    
    samtools sort -T $Input_DIR/tmp/aln.sorted -o $SUB_NAME.sorted.bam $BAM
    echo 'indexing'
    echo $SUB_NAME    
    samtools index $SUB_NAME.sorted.bam
    echo 'next'    
done

```

```bash
## pileup
for BAM  in `ls  $Input_DIR/*.sorted.bam`

do
    TMP_NAME=`basename $BAM`
# 	echo $TMP_NAME
	SUB_NAME=`echo $TMP_NAME |awk -F '.bai' '{print $1}'`
	echo $SUB_NAME
    bcftools mpileup  --fasta-ref $hg19 -o $SUB_NAME.sorted.bam.mp $BAM
    echo 'next'    
done

```

```bash
## call
for BAM  in `ls  $Input_DIR/*.mp`

do
    TMP_NAME=`basename $BAM`
# 	echo $TMP_NAME
	SUB_NAME=`echo $TMP_NAME |awk -F '.mp' '{print $1}'`
	echo $SUB_NAME
    bcftools call -c -o $SUB_NAME.vcf $BAM 

    echo 'next'    
done
```
